```python
import pandas as pd
names = ['preg', 'plas', 'pres', 'skin', 'test','mass', 'pedi', 'age', 'class']
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv'
data = pd.read_csv(url, names=names)
data
```

```python
correlation = data.corr(method='pearson') # skin & test corr but you take out skin 0.07 bc is correlated but also small (0.07) when related to diabetes (class)
correlation
```

```python
data.skew() # sign determines the side to which the biggest bump is
# zero is centered, gauss perfect 
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,10))
ax = fig.gca()
data.hist(ax=ax, color='powderblue') # also density = probability
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.gca()
data.plot(ax=ax, kind='density', subplots=True, layout=(3,3))
plt.show()
```

```python
correlation = data.corr(method='pearson')
fig2 = plt.figure()
ax=fig2.add_subplot()
cax=ax.matshow(correlation, vmin=-1, vmax=1)
fig2.colorbar(cax)
ticks = np.arange(0,9)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```
